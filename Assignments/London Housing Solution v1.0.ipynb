<b>LONDON HOUSING CASE- UNIT 4 </b>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col=0, parse_date=True)
properties.head()
```

<b>Delete MISSING VALUES & UNWANTED INFO</b>

```python
properties.dropna(axis=1, inplace=True)
df = properties.transpose().reset_index()
df.rename(columns={'index':'London Borough'}, inplace=True)
df.columns = df.columns.fillna('ID')
df = df.drop(columns=['ID']).set_index('London Borough')
df.head()
```

<b>Set DATETIMEINDEX</b>

```python
df = df.transpose()
df.index = pd.DatetimeIndex(df.index)
df.head()
```

<b>Filter for exactly 32 Boroughs</b>

```python
df = df.drop(df.columns[-12:], axis=1)
df = df.drop(columns=['City of London'])
df.info()
```

<b>Convert value to FLOAT data type for statistical operations</b>

```python
df = df.astype('float')
df.head()
```

<b>DOWNSAMPLE data to mean annual housing price in each borough</b>

```python
df = df.resample('A').mean()
df.head()
```

<b>Calculating annual % change in each borough</b>

```python
br = df.pct_change() *100
br.head()
```

<b>FINAL RESULT - Calculate overall average % increase of housing price in each borough</b>

```python
result = br.mean().sort_values(ascending=False)

#Top 5 boroughs with greatest increase in housing price
result[:5]
```

<b>Line Plot indicating yearly % change of housing price in Hackney, i.e, the borough with Highest increase!</b>

```python
br['Hackney'].plot(figsize=(15,7), title='Hackney- Annual change in housing price')
```

<b>Bar Plot displaying the Average increase of Housing price in 32 London Boroughs since 1995</b>

```python
tit = 'Average Increase (%) of Housing price in 32 London Boroughs since 1995'
result.plot(kind='bar', title=tit, figsize=(20,7), fontsize=12, ylim=(7,10))
```

<b>CONCLUSION</b>

<b>HACKNEY has highest average housing price increase at 9.43%</b>
